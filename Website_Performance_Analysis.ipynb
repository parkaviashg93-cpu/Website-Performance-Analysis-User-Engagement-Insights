**Dataset Columns / Features Description:**

- **Session primary channel group:** The marketing channel (e.g., Direct, Organic Social)  

- **Date + hour (YYYYMMDDHH):** The specific date and hour of the session  

- **Users:** Number of users in a given period  

- **Sessions:** Number of sessions in that period  

- **Engaged sessions:** Number of sessions with significant user engagement  

- **Average engagement time per session:** The average time a user is engaged per session  

- **Engaged sessions per user:** Ratio of engaged sessions to total sessions per user  

- **Events per session:** Average number of events (actions taken) per session  

- **Engagement rate:** The proportion of sessions that were engaged  

- **Event count:** Total number of events during the period  

### Analysis

- The analysis shows that user traffic and sessions vary across different marketing channels, with some channels contributing more to overall engagement than others.

- Engagement levels differ significantly based on user activity, where higher sessions do not always result in higher engagement rates, indicating variation in user interaction quality.

- Time-based analysis (date and hour) reveals fluctuations in user activity, showing peak usage periods and low-traffic intervals.

- Some sessions have low engagement despite high event counts, suggesting that users perform actions but may not stay engaged for longer durations.

# Check Skewness & Apply Transformations

### Check skewness of numeric columns

```python
df.skew(numeric_only=True)
```

### Apply Transformation

```python
df['avg_engagement_time'] = np.log1p(df['avg_engagement_time'])
df['events_per_session'] = np.log1p(df['events_per_session'])
df[['avg_engagement_time', 'events_per_session']].skew()
```

## **Check Skewness and Apply Transformations**

- Checked skewness of numerical columns using `.skew()`  
- Observed high positive skewness in **avg_engagement_time** and **events_per_session**  
- Applied log transformation using `np.log1p()` to reduce skewness  
- After transformation, skewness was significantly reduced  
- Other columns had acceptable skewness, so no transformation was applied  

# Convert Data Types (if needed)

### Check Current Data Types

```python
df.dtypes
```

### Convert date_hour → datetime

```python
df['date_hour'] = pd.to_datetime(df['date_hour'], format='%Y%m%d%H')
```

### Convert channel_group → category

```python
df['channel_group'] = df['channel_group'].astype('category')
```

```python
df.dtypes
```

## **Convert Data Types**

- Checked data types of all columns using `.dtypes`  
- Converted **date_hour** from integer format to datetime using `pd.to_datetime()`  
- Converted **channel_group** from object to categorical data type  
- Other columns already had appropriate numeric data types, so no further conversion was required  

# Feature Transformation

### Extract date parts

```python
# Extract date parts
df['year'] = df['date_hour'].dt.year
df['month'] = df['date_hour'].dt.month
df['day'] = df['date_hour'].dt.day
df['hour'] = df['date_hour'].dt.hour
df['day_of_week'] = df['date_hour'].dt.day_name()
```

### Create Useful Derived Features

```python
# Engagement per session (already exists but ensure clarity)
df['engagement_per_session'] = df['engaged_sessions'] / df['sessions']
```

```python
# Events per user (new useful metric)
df['events_per_user'] = df['event_count'] / df['users']
```

```python
df.head()
```

## **Feature Transformation**

- Extracted date-related features such as **year, month, day, hour, and day_of_week** from the **date_hour** column  
- Created new derived features like **engagement_per_session** and **events_per_user** for better analysis  
- These features help in understanding user behavior patterns and website performance trends more effectively  

# Stage 3 – EDA and Visualizations

### UNIVARIATE ANALYSIS

### Sessions

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Sessions'], kde=True)
plt.title('Distribution of Sessions')
plt.xlabel('Sessions')
plt.ylabel('Frequency')
plt.show()
```

**Insights:** Sessions are mainly distributed between 20–80 with highest concentration around 40–60 and very few values extending beyond 150 up to ~300.

**Interpretation:**
- Shows distribution of Sessions  
- Majority values lie between 20–80  
- Peak concentration around 40–60  
- Very few values extend beyond 150 up to ~300  
- X-axis → Sessions, Y-axis → Frequency  
- Indicates most users have moderate activity with very few highly active users  

###  Engagement Time

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=np.log1p(df['Average engagement time per session']))
plt.title('Log Transformed Engagement Time')
plt.show()
```

**Insights:** Log-transformed engagement time is mostly concentrated between ~3.5–4.5 with fewer extreme values indicating normalized distribution after reducing skewness.

**Interpretation:**
- Shows distribution of log-transformed engagement time  
- Most values lie between ~3.5–4.5  
- Fewer extreme outliers compared to original data  
- Y-axis → Log Engagement Time  
- Indicates skewness is reduced and data is more evenly distributed  

### Traffic Channel

```python
plt.figure(figsize=(6,4))
sns.countplot(x=df['Session primary channel group (Default channel group)'])
plt.title('Traffic Channel Distribution')
plt.xlabel('Channel')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**Insights:** Direct, Organic Social, Organic Search, and Referral channels have highest counts (670–680) while Unassigned (370), Organic Video (120), and Email (5) have significantly lower counts.

**Interpretation:**
- Shows distribution of traffic channels  
- Direct, Organic Social, Organic Search, and Referral dominate (670–680)  
- Unassigned is moderate (370) while Organic Video (~120) and Email (5) are minimal  
- X-axis → Channel, Y-axis → Count  
- Indicates majority traffic comes from top 4 channels with very low contribution from Email and Video

### Correlation Heatmap of Website Performance Metrics

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Website Performance Metrics')
plt.show()
```

**Insights:** Users, Sessions, Engaged sessions, and Event count show very strong positive correlation (~0.95–0.99) while Average engagement time per session shows weak or negative correlation with them.

**Interpretation:**
- Shows correlation among website performance metrics  
- Strong positive correlation (~0.95–0.99) between Users, Sessions, Engaged sessions, and Event count  
- Average engagement time per session shows weak/negative correlation (~-0.09 to 0.07)  
- Engagement rate strongly correlates with Engaged sessions per user (~0.96)  
- Color intensity (red = strong positive, blue = weak/negative)  
- Indicates traffic-related metrics move together while engagement time behaves differently

- Indicates Organic Social drives highest user activity while Email and Video contribute minimal traffic

### Advanced Plot – Line Chart (Time vs Sessions)

```python
# Step 1: Convert to datetime
df['Date + hour (YYYYMMDDHH)'] = pd.to_datetime(df['Date + hour (YYYYMMDDHH)'], format='%Y%m%d%H')

# Step 2: Group by time
time_data = df.groupby('Date + hour (YYYYMMDDHH)')['Sessions'].mean().reset_index()

# Step 3: Plot
plt.figure(figsize=(8,5))

sns.lineplot(x=time_data['Date + hour (YYYYMMDDHH)'], y=time_data['Sessions'])

plt.title('Sessions Trend Over Time')
plt.xlabel('Time')
plt.ylabel('Sessions')
plt.xticks(rotation=45)
plt.show()
```

**Insights:** Sessions fluctuate mostly between ~20–80 with periodic spikes reaching ~100–110 indicating consistent but varying user activity over time.

**Interpretation:**
- Shows trend of Sessions over time  
- Most values lie between ~20–80  
- Regular fluctuations with peaks around ~100–110  
- X-axis → Time, Y-axis → Sessions  
- Indicates steady user activity with periodic increases in traffic
